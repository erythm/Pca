## Exercise 2. Analyzing the Variance and the PCs

In the cells below, do the following operations:
1. create two new dataframes from _X_df_ applying a StandardScaler and a MinMaxscaler. Call these new dataframes as _Xstd_df_ and _Xmm_df_, respectively;
2. compute the variance of all the features in _X_df_, _Xstd_df_, and _Xmm_df_ and **comment the results**;
3. compute all the $n$ Principal Components (PCs) for each dataset _X_df_, _Xstd_df_, and _Xmm_df_. Then, visualize the curves of the cumulative explained variances and **comment the results**.

```python
#standardization:mean = 0, std = 1
scaler_std = StandardScaler()
xstd = scaler_std.fit_transform(x_df_clean)
xstd_df = pd.DataFrame(xstd, columns=x_df_clean.columns)

#min-max Scaling: scales each feature to [0, 1]
scaler_mm = MinMaxScaler()
xmm = scaler_mm.fit_transform(x_df_clean)
xmm_df = pd.DataFrame(xmm, columns=x_df_clean.columns)
```

```python
#compute variance
var_orig = x_df_clean.var()
var_std = xstd_df.var()
var_mm = xmm_df.var()

print("Mean variance of original data:", var_orig.mean())
print("Mean variance of standardized data:", var_std.mean())
print("Mean variance of MinMax-scaled data:", var_mm.mean())
```

In the original dataset, the variance varies widely among features due to different scales.
After standardization, all features have variance â‰ˆ 1 (as expected).

With MinMax scaling, variance is reduced because all values are forced into [0, 1]; this transformation preserves shape but not variance magnitude.

StandardScaler is giving all features equal importance.

```python
#compute function 
def compute_pca_and_plot(x, label):
    pca = PCA()
    pca.fit(x)
    cum_var = np.cumsum(pca.explained_variance_ratio_)
    
    plt.plot(range(1, len(cum_var)+1), cum_var, label=label)
    return cum_var

plt.figure(figsize=(10,6))
cum_var_orig = compute_pca_and_plot(x_df_clean, 'Original')
cum_var_std = compute_pca_and_plot(xstd_df, 'Standardized')
cum_var_mm = compute_pca_and_plot(xmm_df, 'MinMax Scaled')

plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("Cumulative Explained Variance by PCA")
plt.legend()
plt.grid(True)
plt.show()
```

The Original data (unscaled) curve grows fastest early on â€” because high-variance features dominate.

The Standardized and MinMax-scaled curves grow more gradually and similarly â€” but Standardized is more interpretable and preferred for PCA.

## Exercise 4. $k$-Means

In the cells below, do the following operations:
1. For each one of the two datasets (_std_ and _mm_), run the $k$-Means for clustering the data. In particular, **use the silohuette score for identify the best value for $k\in\{3, \ldots, 10\}$**.
2. Plot the score graphs of exercise 3.3, adding the centroids of the cluster.
3. Observing the centroids coordinates in the PC space, **give a name/interpretation to them**, exploiting the names you assigned to the PCs. **Comment and motivate your interpretations**.

```python
def best_k_by_silhouette(x, label):
    best_k = None
    best_score = -1
    for k in range(3, 11):
        kmeans = KMeans(n_clusters=k, random_state=random_seed)
        labels = kmeans.fit_predict(x)
        score = silhouette_score(x, labels)
        print(f"{label} â€“ k={k}: Silhouette score = {score:.4f}")
        if score > best_score:
            best_k = k
            best_score = score
    print(f"\nBest k for {label}: {best_k} (score = {best_score:.4f})")
    return best_k

best_k_std = best_k_by_silhouette(xstd_pca, "Standardized PCA")
best_k_mm = best_k_by_silhouette(xmm_pca, "MinMax PCA")
```

```python
#cluster with best k for Standardized PCA
kmeans_std = KMeans(n_clusters=best_k_std, random_state=random_seed)
labels_std = kmeans_std.fit_predict(xstd_pca)
centroids_std = kmeans_std.cluster_centers_

#plot
plt.figure(figsize=(8, 6))
plt.scatter(xstd_pca[:, 0], xstd_pca[:, 1], c=labels_std, cmap='tab10', alpha=0.6)
plt.scatter(centroids_std[:, 0], centroids_std[:, 1], color='red', marker='x', s=200, label='Centroids')
plt.xlabel("PC1 â€“ Sociability / Positivity")
plt.ylabel("PC2 â€“ Optimism and Social Reach")
plt.title("k-Means Clustering â€“ Standardized PCA")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#cluster with best k for MinMax PCA
labels_mm = kmeans_mm.fit_predict(xmm_pca)
centroids_mm = kmeans_mm.cluster_centers_

# Plot
plt.figure(figsize=(8, 6))
plt.scatter(xmm_pca[:, 0], xmm_pca[:, 1], c=labels_mm, cmap='tab10', alpha=0.6)
plt.scatter(centroids_mm[:, 0], centroids_mm[:, 1], color='red', marker='x', s=200, label='Centroids')
plt.xlabel("PC1 â€“ Engagement and Activity")
plt.ylabel("PC2 â€“ Sociability and Vitality")
plt.title("k-Means Clustering â€“ MinMax PCA")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Create dataframe for centroid interpretation (Standardized PCA)
centroids_std_df = pd.DataFrame(centroids_std, columns=[f"PC{i+1}" for i in range(xstd_pca.shape[1])])
print("Centroid coordinates (Standardized PCA):")
display(centroids_std_df)
```

We used k-Means clustering (k = 7) on the PCA-reduced data.
By interpreting the cluster centroids in the PC space, we assigned profiles to each group based on sociability, optimism, and emotional reactivity.
The clusters range from Social Optimists (high engagement and stability) to Detached Thinkers (low sociability and ambition but emotionally calm).
This analysis could help companies target content, products, or mental wellness campaigns based on personality profiles.

## Exercise 5. Cluster Evaluations

In the cells below, do the following operations:
1. For each one of the two datasets (_std_ and _mm_), perform an **external evaluation** of the clustering obtained at exercise 4.1 with respect to one or more labels in the list _labels_. **Comment the results, comparing the evaluation with the interpretation you gave at exercise 4.3**. 
2. For each one of the two datasets (_std_ and _mm_), perform an **internal evaluation** of each cluster, with respect to the silohuette score. **Comment the results**. 

```python
from sklearn.metrics import adjusted_mutual_info_score

#reload demographic labels
labels_df = df[labels].copy().loc[x_df_clean.index]

#external eval for each label in the list
for label in labels:
    if labels_df[label].isnull().any():
        continue  # Skip if there are missing values
    ami_std = adjusted_mutual_info_score(labels_df[label], labels_std)
    print(f"AMI between clusters (standardized) and {label}: {ami_std:.4f}")
```

```python
for label in labels:
    if labels_df[label].isnull().any():
        continue
    ami_mm = adjusted_mutual_info_score(labels_df[label], labels_mm)
    print(f"AMI between clusters (MinMax) and {label}: {ami_mm:.4f}")
```

We evaluated the alignment between the clustering results and the demographic label Education using the Adjusted Mutual Information (AMI) score.
For the Standardized PCA, the AMI was 0.0040
For the MinMax PCA, the AMI was 0.0037

These values are extremely close to zero, indicating that the clusters are not aligned with participants' education levels.
This is a desirable result in our context. It suggests that the clusters represent behavioral or psychological profiles, rather than being influenced by demographic categories. This supports the interpretation that the clusters we identified (e.g., Social Optimists, Stressed Socializers) reflect underlying traits or habits, not formal background like education.

```python
# Get silhouette score per sample
sample_silhouette_values = silhouette_samples(xstd_pca, labels_std)

# Compute avg score per cluster
cluster_scores = []
for i in range(best_k_std):
    cluster_score = np.mean(sample_silhouette_values[np.array(labels_std) == i])
    cluster_scores.append(cluster_score)
    print(f"Cluster {i}: Avg silhouette score = {cluster_score:.4f}")
```

```python
#repeat for min_max
sil_mm_total = silhouette_score(xmm_pca, labels_mm)
print(f"Overall silhouette score (MinMax PCA): {sil_mm_total:.4f}")

# Per-cluster silhouette score
sil_mm_sample = silhouette_samples(xmm_pca, labels_mm)
for i in range(best_k_mm):
    score = np.mean(sil_mm_sample[np.array(labels_mm) == i])
    print(f"Cluster {i} (MinMax): Avg silhouette = {score:.4f}")
```
